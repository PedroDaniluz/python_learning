# Métodos lista
list()

### lista.append(elemento)
Adiciona um novo elemento no final da lista
```python
lista = [1, 2, 23, 5, 5, 7, 8]

lista.append(3)
print(f"append(3): {lista}")
# -> append(3): [1, 2, 23, 5, 5, 7, 8, 3]
```

### lista.index(elemento)
Retorna a posição do elemento na lista
```python
lista = [1, 2, 23, 5, 5, 7, 8, 3]

print(f"index(3): {lista.index(3)}")
# -> index(3): 7
```

### lista.insert(elemento, posicao)
Adiciona um elemento em uma posição definida da lista
```python
lista = [1, 2, 23, 5, 5, 7, 8, 3]

lista.insert(4, 5)
print(f"insert(4, 5): {lista}\n")
# -> insert(4, 5): [1, 2, 23, 5, 5, 5, 7, 8, 3]
```

### lista.count(elemento)
Retorna a quantidade de vezes que um elemento apareceu na lista
```python
lista = [1, 2, 23, 5, 5, 5, 7, 8, 3]

print(f"count(5): {lista.count(5)}\n")
# -> count(5): 3
```

### lista.pop()
Remove o último elemento da fila
```python
lista = [1, 2, 23, 5, 5, 5, 7, 8, 3]

lista.pop()
print(f"pop(): {lista}\n")
# -> pop(): [1, 2, 23, 5, 5, 5, 7, 8]
```

### lista.pop(posicao)
Remove o elemento na posição dada como parâmetro
```python
lista = [1, 2, 23, 5, 5, 5, 7, 8]

lista.pop()
print(f"pop(0): {lista}\n")
# -> pop(0): [2, 23, 5, 5, 5, 7, 8]
```

### lista.reverse()
Inverte a ordem dos elementos de uma lista
```python
lista = [2, 23, 5, 5, 5, 7, 8]

lista.reverse
print(f"reverse: {lista}")
# -> reverse: [8, 7, 5, 5, 5, 23, 2]
```

### lista.sort()
Ordena a lista em ordem crescente
```python
lista = [8, 7, 5, 5, 5, 23, 2]

lista.sort(reverse=False)
print(f"sort: {lista}\n")
# -> sort: [2, 5, 5, 5, 7, 8, 23]
```

### lista.sort(reverse=True)
Ordena a lista em ordem decrescente
```python
lista = [2, 5, 5, 5, 7, 8, 23]

lista.sort(reverse=True)
print(f"sort reverse: {lista}\n")
# -> sort reverse: [23, 8, 7, 5, 5, 5, 2]
```
